```python
# Importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df =pd.read_csv("2.Teachers report final.csv")

df['Needs Attention %'] = df['Needs Attention %'] * 100
df['Satisfactory %'] = df['Satisfactory %'] * 100
df['Unsatisfactory %'] = df['Unsatisfactory %'] * 100
# 1. Basic Information

print("Basic Information:")
print(df.info())

# 2. Descriptive Statistics
print("\nDescriptive Statistics:")
print(df.describe())

# 3. Distribution Analysis
# Histograms for percentage columns
df[['Satisfactory %', 'Needs Attention %', 'Unsatisfactory %']].hist(bins=5, figsize=(10, 5))
plt.suptitle('Distribution of Percentages', fontsize=16)
plt.show()

# Box plots for numerical data
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[['Satisfactory %', 'Needs Attention %', 'Unsatisfactory %']])
plt.title('Box Plot of Percentage Metrics', fontsize=16)
plt.show()

# 4. Correlation Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(df[['Satisfactory %', 'Needs Attention %', 'Unsatisfactory %']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap', fontsize=16)
plt.show()

# 5. Comparing Categories (Stacked Bar Plot)
df.set_index('SUB DOMAIN')[['Satisfactory %', 'Needs Attention %', 'Unsatisfactory %']].plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Comparison of Metrics across Sub Domains', fontsize=16)
plt.ylabel('Percentage')
plt.xticks(rotation=45, ha='right')
plt.show()

# 6. Top/Bottom Performing Areas
print("\nTop Performing Sub Domains (Satisfactory %):")
print(df[['SUB DOMAIN', 'Satisfactory %']].sort_values(by='Satisfactory %', ascending=False).head())

print("\nBottom Performing Sub Domains (Needs Attention %):")
print(df[['SUB DOMAIN', 'Needs Attention %']].sort_values(by='Needs Attention %', ascending=False).head())
```
